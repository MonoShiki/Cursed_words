```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests
import pandas as pd
import numpy as np
```

```python
html = requests.get("https://proza.ru/2016/08/29/1210")
soup = bs(html.text)
p1 = [st for st in soup.find_all(name="div",attrs='text')[0].text.strip().split('\n')[1:] if len(st.split()) > 1]
p2 = [st.split("-")[:1][0].split("(")[:1][0].lower().strip() for st in p1]
off_w = set()
for st in p2:
    if ":"  in st or "слова" in st or st == "без" or "давить" in st or "от," in st:
        continue
    if "," not in st and len(st.split()) > 1:
        continue
    l = st.split("—")[0].split("–")[0].strip()
    if len(l.split()) > 1 and "," not in l:
        continue
    for w in l.split(","):
        if w:
            off_w.add(w.strip())
pprint(off_w)
ds = pd.Series(data= list(map(str,off_w)),name= "Offencive_Words")
ds.to_csv("off_words.csv")
```
